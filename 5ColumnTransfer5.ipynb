```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('/content/covid_toy.csv')
```

```python
df.sample(5) #as we can see that gender,city is a nominal and cough is ordinal and on has_covid label encoder
```

```python
df.isnull().sum() #as we can see that there is missing values in fever column
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

#1.Aam Jindagi (converting the columns one by one)

```python
# filling the missing values of the data with the mean of the data by using simple imputer
si= SimpleImputer()
x_train_fever=si.fit_transform(x_train[['fever']])
x_test_fever=si.transform(x_test[['fever']])
x_train_fever.shape
```

```python
#Ordinal Encoding on cough
oe= OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough=oe.fit_transform(x_train[['cough']])
x_test_cough=oe.transform(x_test[['cough']])
x_train_cough.shape
```

```python
#one hot encoding -> gender,city
ohe= OneHotEncoder(drop='first',sparse=False) #we use sparse as we are passing two of the columns and drop ='first' as machine understood the default value so it helps to reduce the columns
x_train_gender_city=ohe.fit_transform(x_train[['gender','city']])
x_test_gender_city=ohe.transform(x_test[['gender','city']])
x_test_gender_city.shape
```

```python
#now accesing the age column
x_train_age=x_train.drop(columns=['gender','city','fever']).values
x_test_age=x_test.drop(columns=['gender','city','fever']).values
x_test_age.shape
```

```python
x_train_transformed=np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis=1) #total 7 columns will be formed but cough creates a numpy wit its own string column
x_train_transformed=np.concatenate((x_test_age,x_test_fever,x_test_gender_city,x_test_cough),axis=1)
x_train_transformed
```

```python
final=pd.DataFrame(x_train_transformed)#we have converted the numpy into the dataframe to drop the mild column
```

```python
final=final.drop(columns=[1])
```

# Column transfer for doing the above task easily

```python
from sklearn.compose import ColumnTransformer
transformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(drop='first',sparse=False),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(x_train)
transformer.transform(x_test)
```
