```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/customer.csv')
```

```python
df.sample(5)
```

```python
 df=df.iloc[:, 2:]
df.head()
```

as we know before doing feature scaling we always do test train split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2) #in x we pass the input column in y we pass the output column
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']]) #passing categories in the increasing priority order
```

```python
oe.fit(x_train)
```

```python
x_train=oe.transform(x_train)
x_test=oe.transform(x_test)
```

label encoding for the output column

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()  #in label encoding we cant assign the priority
```

```python
le.fit(y_train) #we always fit the training dataset and transform both train and test
```

```python
y_train=le.transform(y_train)
y_test=le.transform(y_test)
```
