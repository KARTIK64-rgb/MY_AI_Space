```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape   # it has 60k photos in 28*28 picxels
```

```python
x_train[0]   # we will convert the array in range of 0to 1 so will will divide each of the no. with the maximum value 255
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
```

```python
x_train=x_train/255
x_test=x_test/255
x_train.shape
```

```python
x_train[0]
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train.reshape(1, 28*28), y_train)
```

```python
y_pred=lr.predict(x_test.reshape(-1, 28*28))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train.reshape(-1, 28*28), y_train)
```

```python
y_pred2=knn.predict(x_test.reshape(-1, 28*28))
```

```python
accuracy_score(y_test, y_pred2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train.reshape(-1, 28*28), y_train)

y_pred3=rf.predict(x_test.reshape(-1, 28*28))
accuracy_score(y_test, y_pred3)
```
